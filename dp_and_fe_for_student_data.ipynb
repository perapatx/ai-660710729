##Mean Replcing

```python
df.isnull().sum()
#สังเกตได้ว่า คอลัมน์ที่เป็นตัวเลข และสามารถหาค่าเฉลี่ยได้ คืออ อายุ เกรด ส่วนสูง น้ำหนัก และชั่วโมงเรียนต่อสัปดาห์
```

mean replace คือแทนค่าที่มี missing ด้วยค่าเฉลี่ย ด้วยฟังก์ชั่น fillna

```python
#หาค่าเฉลี่ยอายุ ส่วนสูง น้ำหนัก และชั่วโมงเรียนต่อสัปดาห์
mean_age = df['อายุ'].mean()
mean_grade = df['GPA'].mean()
mean_height = df['ส่วนสูง_cm'].mean()
mean_weight = df['น้ำหนัก_kg'].mean()
mean_hour = df['ชั่วโมงเรียน_ต่อสัปดาห์'].mean()

#แทนค่าที่มี missing ด้วยค่าเฉลี่ย ด้วยฟังก์ชั่น fillna
df['อายุ'].fillna(mean_age, inplace=True)
df['GPA'].fillna(mean_grade,inplace=True)
df['ส่วนสูง_cm'].fillna(mean_height, inplace=True)
df['น้ำหนัก_kg'].fillna(mean_weight, inplace=True)
df['ชั่วโมงเรียน_ต่อสัปดาห์'].fillna(mean_hour, inplace=True)

df
#จะได้ว่า ค่า Missing value ที่เป็นตัวเลข ได้ถูกแก้ไขเร้บร้อย
```

##Standarization (ข้อมูลที่เป็นมาตรฐาน )

```python
#คือข้อมูลบางข้อมูล เช่น เพศ มีการใส่ M,Male,ชาย มา ไม่เหมือนกัน ดังนั้น จึงต้องแก้
df
#ในที่นี้มี เพศ และ คณะ
```

คือข้อมูลบางข้อมูล เช่น เพศ มีการใส่ M,Male,ชาย มา ไม่เหมือนกัน ดังนั้น จึงต้องแก้ ด้วยการสร้าง เมธอด ขึ้นมากรอง

```python
#คณะ มีข้อมูลอะไรบ้างd
print(df['คณะ'])
print(df['เพศ'])
```

จะมีคณะ วิทยาศาสตร์ - วิทย์ , Science ,Sci
 วิศวกรรมศาสตร์ - วิศวกรรม,วิศวะ ,Engineering ,วิศวฯ
 ศิลปศาสตร์ - Arts,ศิลป์ ,Liberal Arts
 และ
 เพศ ชาย - Male ,M,ช
 หญิง - Female,F,ญ

```python
#สร้าง ฟังก์ชั่นมากำกับคำ
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์','Sci']: #วิทยาศาสตร์ - วิทย์ , Science ,Sci
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts','Liberal Arts']:#ศิลปศาสตร์ - Arts,ศิลป์ ,Liberal Arts
        return 'ศิลปศาสตร์'
    elif name in ['วิศวกรรมศาสตร์','วิศวกรรม','วิศวะ','Engineering','วิศวฯ']: #วิศวกรรมศาสตร์ - วิศวกรรม,วิศวะ ,Engineering ,วิศวฯ
        return 'วิศวกรรมศาสตร์'
    else:
        return name

def standardize_gender(name):
    if name in ['ชาย','Male','M','ช']:
        return 'ชาย'
    elif name in ['หญิง','Female','F','ญ']:
        return 'หญิง'
    else:
        return name

#แทนที่ คณะแบบเียนผิดๆถูกๆ
df['คณะ'] = df['คณะ'].apply(standardize_faculty)
df['เพศ'] = df['เพศ'].apply(standardize_gender)
df
```

##Abnormaly Value

```python
df
#ในที่นี่ เห็นอายุ และ gpa
```

คือการจัดการค่าท่ีมาก น้อย ผิดปกติ ในที่นี้มี อายุ และ เกรด อายุ จะใช้การหา Quantile เป็นขอบเขต ส่วน เกรด จะใช้ 1 และ 4 เป็นขอบเขต

```python
#Age
Q1_Age = df['อายุ'].quantile(0.25)
Q3_Age = df['อายุ'].quantile(0.75)
IQR_Age = Q3_Age - Q1_Age
#ขอบเขตของ อายุ
lower_bound_age = Q1_Age - 1.5 * IQR_Age
upper_bound_age = Q3_Age + 1.5 * IQR_Age
print(lower_bound_age,upper_bound_age)

#gpa
lower_bound_gpa = 1
upper_bound_gpa = 4
print(lower_bound_gpa,upper_bound_gpa)
```

```python
#กรองค่าผิดปกติออก ของอายุ และเกรด
df_no_outliers = df[((df['อายุ'] >= lower_bound_age) & (df['อายุ'] <= upper_bound_age )) & ((df['GPA'] >= 1) & (df['GPA'] <= 4))]

#df_no_outliers = df_abnormal[(df_abnormal['GPA'] >= lower_bound_gpa) & (df_abnormal['GPA'] <= upper_bound_gpa)]
df = df_no_outliers
#ข้อมูลที่ อายุ เกินไป กับ เกรด น้อยกว่า 1 มากกว่า 4 จะถูกตัดไป
df_no_outliers
```
